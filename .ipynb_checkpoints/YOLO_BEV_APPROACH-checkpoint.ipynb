```python
import open3d as o3d
import os
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
import plotly.express as px

from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load PCD file
pcd_path = 'label_cloud_project/pointclouds_cropped/cropped_40.pcd'
point_cloud = o3d.io.read_point_cloud(pcd_path)

# Convert point cloud to NumPy array
points = np.asarray(point_cloud.points)
```

## Crop Point Clouds

def crop_point_cloud(pcd_path, distance_threshold):
    # Load PCD file
    point_cloud = o3d.io.read_point_cloud(pcd_path)
    
    # Convert point cloud to NumPy array
    points = np.asarray(point_cloud.points)
    
    # Calculate Euclidean distance from the origin
    distances = np.linalg.norm(points, axis=1)
    
    # Filter points within the distance threshold
    cropped_points = points[distances <= distance_threshold]
    
    # Create a new Open3D point cloud
    cropped_pcd = o3d.geometry.PointCloud()
    cropped_pcd.points = o3d.utility.Vector3dVector(cropped_points)
    
    return cropped_pcd

# Folder containing PCD files
folder_path = 'label_cloud_project/pointclouds/'
output_folder = 'label_cloud_project/pointclouds_cropped/'  # Specify the output folder

# Euclidean distance threshold 'd'
distance_threshold = 15.0  # Adjust this value as needed

# Ensure the output folder exists, if not, create it
os.makedirs(output_folder, exist_ok=True)

# Crop all PCD files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".pcd"):
        pcd_path = os.path.join(folder_path, filename)
        cropped_pcd = crop_point_cloud(pcd_path, distance_threshold)
        
        # Save cropped PCD file to the specified output folder
        output_filename = os.path.join(output_folder, f"cropped_{filename}")
        o3d.io.write_point_cloud(output_filename, cropped_pcd)

## Plot Point Cloud in 3D

```python
sensor_range = 15 # Define the range around the sensor's origin

# Calculate the Euclidean distance from the origin for each point
distances = np.linalg.norm(points, axis=1)

# Filter points within the sensor's range
valid_points = points[distances <= sensor_range]

# Create a 3D scatter plot using Plotly for valid points within range
fig = go.Figure(data=[go.Scatter3d(
    x=valid_points[:, 0],
    y=valid_points[:, 1],
    z=valid_points[:, 2],
    mode='markers',
    marker=dict(size=1)  # Adjust marker size as needed
)])

# Set layout options for the plot
fig.update_layout(
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z'),
        camera=dict(up=dict(x=0, y=0, z=0))  # Adjusting camera orientation
    ),
    title=f'3D Scatter Plot of Point Cloud Data within {sensor_range} units from Sensor Origin'
)

# Show the plot
fig.show()
```

The cones look extremely tall, even though they aren't. It's just that split apart in this format, the dimension are stretched.

## Plot Point Cloud as 2D color-coded image

```python
# Extract X, Y, Z coordinates
x = points[:, 0]
y = points[:, 1]
z = points[:, 2]

# Define the grid size and create 2D histograms for X, Y axes
x_bins = np.linspace(np.min(x), np.max(x), 100)
y_bins = np.linspace(np.min(y), np.max(y), 100)

hist, x_edges, y_edges = np.histogram2d(x, y, bins=(x_bins, y_bins))

# Normalize the histogram values for better visualization
hist_normalized = np.log(hist + 1)  # Apply logarithm for better contrast

# Create a scatter plot to represent Z-axis intensity as color
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=z, cmap='viridis', s=5, alpha=0.6)
plt.colorbar(label='Z-axis Intensity')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bird\'s Eye View of Point Cloud with Z-axis Intensity')
plt.show()
```

```python
x = points[:, 0]
y = points[:, 1]
z = points[:, 2]

# Create a scatter plot using Plotly
fig = px.scatter(x=x, y=y, color=z, color_continuous_scale='viridis', opacity=0.6,
                 labels={'color': 'Z-axis Intensity'}, title='Bird\'s Eye View of Point Cloud with Z-axis Intensity')
fig.update_traces(marker=dict(size=5))

# Set axis labels
fig.update_xaxes(title='X-axis')
fig.update_yaxes(title='Y-axis')

# Calculate the aspect ratio based on the range of X and Y coordinates
x_range = np.max(x) - np.min(x)
y_range = np.max(y) - np.min(y)
aspect_ratio = y_range / x_range

# Set the aspect ratio manually
fig.update_layout(yaxis=dict(scaleanchor="x", scaleratio=aspect_ratio))

# Show the plot
fig.show()
```
